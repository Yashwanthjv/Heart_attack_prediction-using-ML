```python


import pandas as pd
import matplotlib.pyplot as plt

path ="C:\\Users\\Vijayalakshmi K V\\OneDrive\\Desktop\\project\\Heart attack prediction\\heart.csv"

data=pd.read_csv(path)
print(data)
print(data.info())


#extracting dependent and independent variables
inputs= data.drop('output','columns')
output= data.drop(['age','sex','cp','trtbps','chol','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall'],'columns')
print("INPUTS:\n\n",inputs)
print("OUTPUT:\n\n",output)


#splitting the data into training set and test set
import sklearn
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test = train_test_split(inputs,output,test_size=0.2,random_state=2)        

print("x_train\n",x_train)
print("x_test\n",x_test)
print("y_train\n",y_train)
print("y_test\n",y_test)

from sklearn.linear_model import LogisticRegression
model= LogisticRegression()  #initializing the model
model.fit(x_train,y_train)   #training

#predicting the result

y_pred = model.predict(x_test)

print("x_test\n",x_test)
print("y_pred\n",y_pred)
print("y_test\n",y_test)


#confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
print(cm)

#Accuracy 

acc=model.score(inputs,output)
print("acc=",acc)


import numpy as np

input_data=np.array([[63,1,3,145,233,1,0,150,0,2.3,0,0,1]])
predicted_result = model.predict(input_data)
print("predicted result=",predicted_result)

if(predicted_result[0]==0):
    print("person does not have any symptoms of heart attack")
else:
    print("the person may get heart attack")

```
